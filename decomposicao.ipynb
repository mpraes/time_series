### Script para decompor a série temporal e descobrir seus componentes: Tendencia, Sazonalidade e Resíduos.
#### Neste caso são aplicados modelos aditivo e multiplicativo

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = [16,8] 
```

```python
df = pd.read_excel('base_volume_portos.xlsx')
```

```python
df_2 = df[df['PLANTA_PROCESSO'] == "1068 - Ponta da Madeira - DESCARGA"]
```

```python
df_g = df_2.resample(rule='M', on='DATA')['VALOR'].sum()
```

```python
titulo_grafico = df_2['PLANTA_PROCESSO'].unique()[0]
```

```python
plt.plot(df_g)
plt.title("Volume Mensal no período de 2017 a 2022 | " + titulo_grafico, fontsize=20)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado_mult = seasonal_decompose(df_g, model="multiplicative")
```

```python
resultado_mult.plot()
plt.show()
```

```python
resultado_add = seasonal_decompose(df_g, model="aditive")
```

```python
resultado_add.plot()
plt.show()
```
