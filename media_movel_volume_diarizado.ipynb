```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = [16,8]

```

```python
df = pd.read_excel('serie_historica_volume_portos.xlsx')
planta = input("Qual planta e processo para escolher: ")
```

```python
df_2 = df[df['PLANTA_PROCESSO'] == planta]
df_2.sort_values('DATA')
```

```python
df_g = df_2.resample(rule='D', on='DATA')['TOTAL_VALOR'].sum()
```

```python
plt.plot(df_g)
plt.title("Volume diario no perÃ­odo de 2017 a 2022 | " + planta, fontsize=20)
plt.show()
```

```python
df_ma = df_g.rolling(window=90).mean()
```

```python
df_ma.sample(10)
```

```python
df_ma.dropna(inplace=True)
```

```python
plt.plot(df_ma)
plt.title("Volume diario no perÃ­odo de 2017 a 2022 | ", fontsize=20)
plt.show()
```
